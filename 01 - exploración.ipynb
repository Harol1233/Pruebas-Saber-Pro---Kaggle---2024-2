```python
# Importamos las libreras necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train_data = pd.read_csv('train.csv')
# Visualizacin de la cantidad de valores faltantes por columna
# Esto nos ayuda a identificar qu columnas tienen datos incompletos
plt.figure(figsize=(10, 6))
train_data.isnull().sum().plot(kind='bar')
plt.title('Cantidad de Valores Faltantes por Columna')
plt.ylabel('Nmero de valores faltantes')
plt.show()

# Visualizacin de la distribucin de 'RENDIMIENTO_GLOBAL'
# Nos permite ver cuntos registros hay en cada categora de rendimiento
plt.figure(figsize=(8, 5))
sns.countplot(x='RENDIMIENTO_GLOBAL', data=train_data)
plt.title('Distribucin de Rendimiento Global')
plt.show()

# Visualizacin de la distribucin de los perodos en los datos
# Observamos la cantidad de registros en cada perodo
plt.figure(figsize=(10, 6))
sns.countplot(x='PERIODO', data=train_data)
plt.title('Distribucin de Perodos')
plt.xticks(rotation=45)  # Rotamos etiquetas para mejor lectura
plt.show()

# Visualizacin de la relacin entre 'RENDIMIENTO_GLOBAL' y 'ESTU_PRGM_ACADEMICO'
# Observamos cmo se distribuye el rendimiento en diferentes programas acadmicos
plt.figure(figsize=(12, 8))
sns.countplot(
    y='ESTU_PRGM_ACADEMICO', 
    hue='RENDIMIENTO_GLOBAL', 
    data=train_data, 
    order=train_data['ESTU_PRGM_ACADEMICO'].value_counts().index  # Ordenamos por frecuencia
)
plt.title('Rendimiento Global por Programa Acadmico')
plt.show()

# Visualizacin de la relacin entre 'RENDIMIENTO_GLOBAL' y 'ESTU_HORASSEMANATRABAJA'
# Para ver si existe alguna correlacin entre el rendimiento y las horas de trabajo semanales
plt.figure(figsize=(10, 6))
sns.countplot(
    x='ESTU_HORASSEMANATRABAJA', 
    hue='RENDIMIENTO_GLOBAL', 
    data=train_data, 
    order=train_data['ESTU_HORASSEMANATRABAJA'].value_counts().index  # Ordenamos por frecuencia
)
plt.title('Rendimiento Global en funcin de Horas de Trabajo Semanales')
plt.xticks(rotation=45)  # Rotamos etiquetas para legibilidad
plt.show()
```
