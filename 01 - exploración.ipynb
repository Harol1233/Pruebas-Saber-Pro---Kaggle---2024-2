```python
# Importamos las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train_data = pd.read_csv('train.csv')
# Visualización de la cantidad de valores faltantes por columna
# Esto nos ayuda a identificar qué columnas tienen datos incompletos
plt.figure(figsize=(10, 6))
train_data.isnull().sum().plot(kind='bar')
plt.title('Cantidad de Valores Faltantes por Columna')
plt.ylabel('Número de valores faltantes')
plt.show()

# Visualización de la distribución de 'RENDIMIENTO_GLOBAL'
# Nos permite ver cuántos registros hay en cada categoría de rendimiento
plt.figure(figsize=(8, 5))
sns.countplot(x='RENDIMIENTO_GLOBAL', data=train_data)
plt.title('Distribución de Rendimiento Global')
plt.show()

# Visualización de la distribución de los períodos en los datos
# Observamos la cantidad de registros en cada período
plt.figure(figsize=(10, 6))
sns.countplot(x='PERIODO', data=train_data)
plt.title('Distribución de Períodos')
plt.xticks(rotation=45)  # Rotamos etiquetas para mejor lectura
plt.show()

# Visualización de la relación entre 'RENDIMIENTO_GLOBAL' y 'ESTU_PRGM_ACADEMICO'
# Observamos cómo se distribuye el rendimiento en diferentes programas académicos
plt.figure(figsize=(12, 8))
sns.countplot(
    y='ESTU_PRGM_ACADEMICO', 
    hue='RENDIMIENTO_GLOBAL', 
    data=train_data, 
    order=train_data['ESTU_PRGM_ACADEMICO'].value_counts().index  # Ordenamos por frecuencia
)
plt.title('Rendimiento Global por Programa Académico')
plt.show()

# Visualización de la relación entre 'RENDIMIENTO_GLOBAL' y 'ESTU_HORASSEMANATRABAJA'
# Para ver si existe alguna correlación entre el rendimiento y las horas de trabajo semanales
plt.figure(figsize=(10, 6))
sns.countplot(
    x='ESTU_HORASSEMANATRABAJA', 
    hue='RENDIMIENTO_GLOBAL', 
    data=train_data, 
    order=train_data['ESTU_HORASSEMANATRABAJA'].value_counts().index  # Ordenamos por frecuencia
)
plt.title('Rendimiento Global en función de Horas de Trabajo Semanales')
plt.xticks(rotation=45)  # Rotamos etiquetas para legibilidad
plt.show()
```
